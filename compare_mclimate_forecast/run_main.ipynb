```python
import yaml
from forecast_processor import process_all_variables
from ar_hazard_index import compute_AR_hazard_index
from summaries import export_summary_csv

# --- Config ---
init_date = "20000101"
source = "reforecast"
server="aware"
domain_name = 'San-Mateo'

domain = {
    "lon_min": -179.5,
    "lon_max": -110.,
    "lat_min": 10.,
    "lat_max": 70.,
}

var_lst = [
    "qpf",
    "ivt",
    "freezing_level",
    "uv",
]

# --- Process Variables ---
fc, final_ds = process_all_variables(
    init_date=init_date,
    var_lst=var_lst,
    domain=domain,
    source=source,
    server=server
)

# --- Compute AR Hazard Index ---
final_ds = compute_AR_hazard_index(
    final_ds
)

print(final_ds)
print(fc)

# --- Load Domain Information ---
with open("config/domain.yaml", "r") as f:
    config = yaml.safe_load(f)
domain_cfg = config["domains"].get(domain_name, config["domains"][domain_name])

# --- Write CSV of Maximum Value within Domain ---
df = export_summary_csv(
    final_ds,
    domain_name,
    domain_cfg,
    init_date,
    mode="max",
)

# # --- Plot ---
# for lead_time in final_ds.lead_time.values:

#     plot_mclimate_forecast_four_panel(
#         final_ds,
#         fc,
#         lead_time,
#         ...
#     )
```
